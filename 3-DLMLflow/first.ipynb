### Quickstart: Compare runs, choose a model, and deploy it to a REST API

In this quickstart, you will:

- Run a hyperparameter sweep on a training script

- Compare the results of the runs in the MLflow UI

- Choose the best run and register it as a model

- Deploy the model to a REST API

- Build a container image suitable for deployment to a cloud platform

```python
## Import necessary libraries
import keras  # Deep learning library
import numpy as np  # Numerical computations
import pandas as pd  # Data handling
from hyperopt import STATUS_OK, Trials, fmin, hp, tpe  # Hyperparameter tuning with Hyperopt
from sklearn.metrics import mean_squared_error  # Evaluation metric
from sklearn.model_selection import train_test_split  # Splitting dataset
import mlflow  # MLflow for experiment tracking
from mlflow.models import infer_signature  # Model signature inference
```

```python
## load the dataset
data=pd.read_csv(
    "https://raw.githubusercontent.com/mlflow/mlflow/master/tests/datasets/winequality-white.csv",
    sep=";",
)
data
```

### Here we want to pass all the feature column as input to ANN model except Quality (feature), Quality is act as output prediction for ANN model , so we try to predict the quality of data

```python
## Split the data into training,validation and test sets
## Split the dataset into training and test sets (75% training, 25% testing)
train,test=train_test_split(data,test_size=0.25,random_state=42)
train
```

```python
##The .ravel() function flattens the array into a 1D array
##original array is [[3], [4], [5]]  converts it to [3, 4, 5]
train[['quality']].values.ravel()
```

```python
train_x=train.drop(['quality'],axis=1).values
train_y=train[['quality']].values.ravel()

## test dataset
test_x=test.drop(['quality'],axis=1).values
test_y=test[['quality']].values.ravel()


## Further split training data into training and validation sets (80% train, 20% validation)
train_x,valid_x,train_y,valid_y=train_test_split(train_x,train_y,test_size=0.20,random_state=42)

## Infer model signature from training data (input/output structure for MLflow logging)
signature=infer_signature(train_x,train_y)
```

```python
np.mean(train_x,axis=0)
```

```python
## Get the number of input features
train_x.shape[1]
```

```python
### ANN Modle
def train_model(params,epochs,train_x,train_y,valid_x,valid_y,test_x,test_y):

    ## Define model architecture
    mean = np.mean(train_x, axis=0)   ## It is calculating mean value of train_x column wise
    var = np.var(train_x,axis=0)

    model = keras.Sequential(
        [
            ## It will give 11 input feature of train_x
            #Specifies that the input will have the same number of features as train_x.
            keras.Input([train_x.shape[1]]),
            # It is Keras preprocessing layers and is used to normalize input data during model training. 
            # It standardizes input features by subtracting the mean and dividing by the standard deviation.
            keras.layers.Normalization(mean=mean, variance=var),
            ## Dense (fully connected) layer with 64 neurons and ReLU activation
            # This is a fully connected (dense) layer
            keras.layers.Dense(64, activation='relu'),
            # Output layer with 1 neuron (since we are predicting a single value)
            keras.layers.Dense(1)
        ]
    )

    # Compile the model
    #Uses Stochastic Gradient Descent (SGD) optimizer with the given learning rate (lr) and momentum (momentum)
    model.compile(optimizer=keras.optimizers.SGD(
        learning_rate=params["lr"],momentum=params["momentum"]
    ),
    loss="mean_squared_error",
    metrics=[keras.metrics.RootMeanSquaredError()]
    )

    ## Train the ANN model with lr and momentum params wwith MLFLOW tracking
    with mlflow.start_run(nested=True):
        model.fit(train_x,train_y,validation_data=(valid_x,valid_y),
                  epochs=epochs,
                  batch_size=64)
        
    ## Evaluate the model
    eval_result=model.evaluate(valid_x,valid_y,batch_size=64)

    eval_rmse=eval_result[1]

    ## Log the parameters and results
    mlflow.log_params(params)
    mlflow.log_metric("eval_rmse",eval_rmse)

    ## log the model

    mlflow.tensorflow.log_model(model,"model",signature=signature)

    return {"loss": eval_rmse, "status": STATUS_OK, "model": model}
```

```python
#Calls train_model using the given params
def objective(params):
    # MLflow will track the parameters and results for each run
    result = train_model(
        params, #A dictionary containing hyperparameters (lr, momentum).
        epochs=3, # Number of times the model will see the full dataset.
        train_x=train_x, #Training data (features and labels).
        train_y=train_y,
        valid_x=valid_x, # Validation data for tuning.
        valid_y=valid_y,
        test_x=test_x, #Test data (not used here but can be useful for evaluation).
        test_y=test_y,
    )
    return result
```

```python
#lr (learning rate): Log-uniform distribution between 1e-5 and 1e-1.
#momentum: Uniform distribution between 0.0 and 1.0.
#These ranges define possible values for tuning.

space={
    "lr":hp.loguniform("lr",np.log(1e-5),np.log(1e-1)), # Learning rate range (log scale)
    "momentum":hp.uniform("momentum",0.0,1.0)  # Momentum range (linear scale)


}
```

```python
mlflow.set_experiment("wine-quality")
## Run hyperparameter tuning using Hyperopt
with mlflow.start_run():
    #Stores trial results.
    trials = Trials()  # Initialize Hyperopt trials object
    best = fmin(
        fn=objective,  # Function to minimize (objective function)
        space=space,  # Hyperparameter space
        algo=tpe.suggest,  # Optimization algorithm
        # Different sets of hyperparameters.
        max_evals=4,  # Number of evaluations (iterations)
        trials=trials  # Store trial results
    )

    # Fetch the details of the best run by short the all trails based on Loss fns
    best_run = sorted(trials.results, key=lambda x: x["loss"])[0]

    # Log the best parameters, loss, and model
    mlflow.log_params(best)
    mlflow.log_metric("eval_rmse", best_run["loss"])
    mlflow.tensorflow.log_model(best_run["model"], "model", signature=signature)
    
    # Print out the best parameters and corresponding loss
    print(f"Best parameters: {best}")
    print(f"Best eval rmse: {best_run['loss']}")
```
